```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt

n = 180  # banyak day
r = 0.0423  # rate of change (laju perubahan)
k = 4.9  # kapasitas maksimal
dt = 0.04  # perubahan waktu per jam
mk = 3.8  # banyak mikroorganisme awal
tmk = 6.1  # mikroorganisme decay awal

mk0 = []
t = []
tmk1 = []


iter = int(n/dt)


for i in range(iter):
    t.append(i * dt)
    mk = mk + r * dt * (1 - mk/k) * mk  # Pertumbuhan mikroorganisme
    mk0.append(mk)
    tmk = tmk + r * dt * (1 - tmk/k) * tmk  # Pembusukan mikroorganisme
    tmk1.append(tmk)


plt.style.use("default")
plt.figure(1, facecolor='#8B4513')  # Set outside background to brown
plt.plot(t, mk0, label="Growth of microorganisms")
plt.plot(t, tmk1, label="Microorganism decay")
plt.legend()
plt.xlabel('Time (days)')
plt.ylabel('Growth (cm)')
plt.title('Growth vs Decay of Microorganisms Over Time')
plt.grid(True)
plt.gca().set_facecolor('#A0522D')  # Set inside plot background to lighter brown
plt.show()
```

kodingan tersebut menjelaskan tentang peningkatan jumlah mikroorganisme dari day ke-3 hingga ke-9 disertai dengan dekomposisi dari nutrisi yang menunjang perkembangan mikroorganisme.
